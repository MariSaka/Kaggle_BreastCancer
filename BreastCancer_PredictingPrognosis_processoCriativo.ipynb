```python
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn import metrics

#import models
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
```

```python
df = pd.read_csv('Breast_Cancer.csv')
df
```

```python
df.describe(include='all')
```

tratamento de dados

```python
df.columns=df.columns.str.capitalize().str.replace(' ','_')
df.head()
```

```python
print(df.columns.values)
```

```python
df['Race'].unique()
df['Race'] = df['Race'].map({'White':1100, 'Black':1010, 'Other':1001})
df.head()
```

```python
df['Marital_status'].unique()
df['Marital_status'] = df['Marital_status'].map({'Married':110000, 'Divorced':101000, 'Single ':100100, 'Widowed':100010, 'Separated':100001})
df.head()
```

```python
df.rename(columns={'T_stage_':'T_stage'},inplace=True)
df['T_stage'].unique()
df['T_stage'] = df['T_stage'].map({'T1':11000, 'T2':10100, 'T3':10010, 'T4':10001})
df.head()
```

```python
df['N_stage'].unique()
df['N_stage'] = df['N_stage'].map({'N1':1100, 'N2':1010, 'N3':1001})
df.head()
```

```python
df['6th_stage'].unique()
df['6th_stage'] = df['6th_stage'].map({'IIA':110000, 'IIIA':101000, 'IIIC':100100, 'IIB':100010, 'IIIB':100001})
df.head()
```

```python
df['Differentiate'].unique()
df['Differentiate'] = df['Differentiate'].map({'Poorly differentiated':11000, 'Moderately differentiated':10100, 'Well differentiated':10010, 'Undifferentiated':10001})
df.head()
```

```python
df['Grade'].unique()
df['Grade'] = df['Grade'].map({'3':1100, '2':1010, '1':1001})
df.head()
```

```python
df['A_stage'].unique()
df['A_stage'] = df['A_stage'].map({'Regional':110, 'Distant':101})
df.head()
```

```python
df['Estrogen_status'].unique()
df['Estrogen_status'] = df['Estrogen_status'].map({'Positive':110, 'Negative':101})
df.head()
```

```python
df['Progesterone_status'].unique()
df['Progesterone_status'] = df['Progesterone_status'].map({'Positive':110, 'Negative':101})
df.head()
```

```python
df['Status'].unique()
df['Status'] = df['Status'].map({'Alive':1, 'Dead':0})
df.head()
```

```python
df.isnull().values.any()
```

```python
df.isnull().sum()
```

```python
BCdata=df.dropna()
BCdata
```

PrediÃ§Ã£o do status do paciente

treinamento do modelo

```python
x = BCdata.drop(['Status'],axis=1)
y = BCdata['Status']
```

```python
#split data
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
#Training model
lr = LinearRegression()
lr.fit(X_train,y_train)
svm = SVR()
svm.fit(X_train,y_train)
rf = RandomForestRegressor()
rf.fit(X_train,y_train)
gr = GradientBoostingRegressor()
gr.fit(X_train,y_train)
```

```python
y_pred1 = lr.predict(X_test)
y_pred2 = svm.predict(X_test)
y_pred3 = rf.predict(X_test)
y_pred4 = gr.predict(X_test)

df1 = pd.DataFrame({'Actual':y_test,'Lr':y_pred1,'svm':y_pred2,'rf':y_pred3,'gr':y_pred4})
df1
```

```python
#Evaluating the Algorithm
score1 = metrics.r2_score(y_test,y_pred1)
score2 = metrics.r2_score(y_test,y_pred2)
score3 = metrics.r2_score(y_test,y_pred3)
score4 = metrics.r2_score(y_test,y_pred4)
print(score1,score2,score3,score4)
```
